```python
import pandas as pd
import numpy as np
import pickle
from google.colab import drive
```

```python
drive.mount('/content/drive')
DATA_PATH = "/content/drive/MyDrive"
infile = open(DATA_PATH+'/visual_tactile_dataset.p','rb')
dataset = pickle.load(infile)
```

```python
#Cria a DataFrame com as 160 colunas dos sensores: 16 para cada uma dos 10 momentos
#Ex: Sensor 12 na label 3: 3_11
column_list = []
for i in range(1,11):
  for j in range(16):
    column_list.append(str(i)+'_'+str(j))

dfexp = pd.DataFrame(columns = column_list)
```

```python
target = []
position = []
for i in dataset:
  for j in dataset[i]:
    for k in dataset[i][j]:
      for l in dataset[i][j][k]:
        #Cria uma DF de uma linha, com o nome do experimento específico
        #Exemplo: Cheez-100_432-right-4
        row_name = str(i)+'-'+str(j)+'-'+str(k)+'-'+str(l)
        row_df = pd.DataFrame(columns = column_list, index=[row_name])

        target.append(str(i))
        position.append(str(k))

        iter = round((dataset[i][j][k][l]['label'][0][2] - dataset[i][j][k][l]['label'][0][0])/9)

        label_list = [dataset[i][j][k][l]['label'][0][0]]

        for p in range(8):
          label_list.append(label_list[p] + iter)

        label_list.append(dataset[i][j][k][l]['label'][0][2])
        
        dataset[i][j][k][l]['label'] = pd.DataFrame(label_list)
        
        for m in range(10):
          label = dataset[i][j][k][l]['label'][0][m]
          for n in range(16):
            row_df.loc[row_name, str(m+1)+'_'+str(n)] = dataset[i][j][k][l]['tactile'].loc[label-1, n]
        #Pega as 3 linhas de tactile referenciadas em label e as coloca na DF de uma linha

        #Concatena a DF de uma linha na DF principal
        dfexp = pd.concat([dfexp, row_df])

dfexp['Position'] = np.array(position)

#Adiciona a coluna target ao DataFrame
targetdf = np.array(target)
dfexp['Target'] = targetdf

dfexp = dfexp.sort_index()
```

```python
#One hot encoding das posições

dfexp = pd.concat([dfexp, pd.get_dummies(dfexp['Position'], prefix='Position',drop_first=True)], axis = 1)
dfexp.drop(['Position'],axis=1, inplace=True)
```
